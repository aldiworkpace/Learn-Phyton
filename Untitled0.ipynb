```python
#upper case dan lower case mengubah huruf besar dan kecil

kalimat = "I Like Drink So much water"
#uppercase
kalimat = kalimat.upper()
print("Contoh Pengguanaan upper : ",kalimat)

#lowercase
example = "NO ONE WILL HELP YOU"
example = example.lower()

print("Conntoh Penggunaan Lower : ",example)
```

```python
#menghapus awalan karakter spasi dalam kalimat (lstrip) = left strip
#menghapus akhiran kalimat karakter spasi (rstrip) = right strip


Example1 = "       Patrick no using strip"
#no using strip
print("NO Using Strip : ",Example1)

#using left strip
Example2 = "        Patrick left strip"

print(" Using Left Strip : ",Example2.lstrip())

#using right strip = rstrip
Example3 = "Patrick Right strip      "

print("Using Right strip : ",Example3.rstrip())

#using strip delete sentences

Example4 = "Smartes Student Is A Patrick"

print("Using strip deleted Sentences : ",Example4.strip("Smartes Student Is"))
```
